```python
from pydub import AudioSegment
import os

# Load the original audio file
audio_file = AudioSegment.from_file("/content/drive/MyDrive/My Work tasks/speaker diarization/collab sample.mp3")

# Output path to store segmented audio files
output_path = "/content/drive/MyDrive/My Work tasks/speaker diarization/"

# Create the output directory if it doesn't exist
os.makedirs(output_path, exist_ok=True)

# Iterate over diarization segments and save each segment to a file
for turn, _, speaker_num in diarization.itertracks(yield_label=True):
    start_time = int(turn.start * 1000)  # convert to milliseconds
    end_time = int(turn.end * 1000)  # convert to milliseconds

    # Create a folder for each speaker if it doesn't exist
    speaker_folder = os.path.join(output_path, f"speaker_{speaker_num}")
    os.makedirs(speaker_folder, exist_ok=True)

    # Extract the segment
    segment = audio_file[start_time:end_time]

    # Define the output file name within the speaker's folder
    output_file = os.path.join(speaker_folder, f"segment_{turn.start:.1f}_{turn.end:.1f}.mp3")

    # Save the segment to the output file
    segment.export(output_file, format="mp3")

    print(f"Segment saved: {output_file}")
```
